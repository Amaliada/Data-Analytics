## Data Wrangling Plan

1. Handle missing values:
   - Drop `review_comment_title`.
   - Create `has_comment` flag for `review_comment_message`.

2. Fix data types:
   - Convert order dates to datetime.
   - Convert payments and product dimensions to numeric.

3. Create derived columns:
   - Purchase date, purchase hour.
   - Weekend/Weekday.

4. Merge datasets:
   - Orders + Customers + Items + Products + Payments + Reviews.

5. Aggregate as needed:
   - Monthly sales.

### Step 1: Handle Missing Values
- Drop `review_comment_title` (too many missing values, not useful).
- Create a flag `has_comment` for `review_comment_message`.


```python
# Drop 'review_comment_title'
reviews = reviews.drop(columns=["review_comment_title"])

# Create flag for comments
reviews["has_comment"] = reviews["review_comment_message"].notnull()
```

### Step 2: Fix Data Types
- Convert order dates to datetime.
- Convert payment values and product dimensions to numeric.


```python
# Convert date columns
date_cols = ["order_purchase_timestamp", "order_approved_at",
             "order_delivered_carrier_date", "order_delivered_customer_date",
             "order_estimated_delivery_date"]

for col in date_cols:
    if col in orders.columns:
        orders[col] = pd.to_datetime(orders[col])

reviews["review_creation_date"] = pd.to_datetime(reviews["review_creation_date"])
reviews["review_answer_timestamp"] = pd.to_datetime(reviews["review_answer_timestamp"])

# Ensure numeric types
payments["payment_value"] = pd.to_numeric(payments["payment_value"], errors="coerce")
for col in ["product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]:
    products[col] = pd.to_numeric(products[col], errors="coerce")
```

### Step 3: Create Derived Columns
- Extract year, month, and day from order_purchase_timestamp.
- Add weekday/weekend flag.


```python
orders["order_purchase_year"] = orders["order_purchase_timestamp"].dt.year
orders["order_purchase_month"] = orders["order_purchase_timestamp"].dt.month
orders["order_purchase_day"] = orders["order_purchase_timestamp"].dt.day
orders["order_purchase_weekday"] = orders["order_purchase_timestamp"].dt.day_name()
orders["is_weekend"] = orders["order_purchase_weekday"].isin(["Saturday", "Sunday"])
```

### Step 4: Merge Datasets
- Join orders with customers, items, products, payments, and reviews.
- Left joins to preserve all orders.


```python
df = (orders
      .merge(customers, on="customer_id", how="left")
      .merge(order_items, on="order_id", how="left")
      .merge(products, on="product_id", how="left")
      .merge(payments, on="order_id", how="left")
      .merge(reviews, on="order_id", how="left")
     )

print("Final merged shape:", df.shape)
df.head()
```

### Step 5: Aggregations
Create some base aggregations to check that the data is ready:
- Monthly sales (total payment_value).
- Average review score per product category.
- Average order value (AOV).


```python
# Monthly sales
monthly_sales = df.groupby(df["order_purchase_timestamp"].dt.to_period("M"))["payment_value"].sum()

# Review score by category
review_by_category = df.groupby("product_category_name")["review_score"].mean().sort_values(ascending=False)

# Average Order Value (AOV)
aov = df.groupby("order_id")["payment_value"].sum().mean()

print("AOV:", round(aov, 2))
monthly_sales.head(), review_by_category.head()
```

## 1) Monthly Sales Trend

**Question:** How do total sales evolve over time?  
**Metric:** Sum of `payment_value` per month.


```python
import matplotlib.pyplot as plt
import pandas as pd

# asigură-te că timestamp-ul este datetime
ts = pd.to_datetime(df["order_purchase_timestamp"], errors="coerce")

monthly_sales = (
    df.assign(order_purchase_timestamp=ts)
      .groupby(df["order_purchase_timestamp"].dt.to_period("M"))["payment_value"]
      .sum()
)

# pentru plot, convertim PeriodIndex -> Timestamp
x = monthly_sales.index.to_timestamp()
y = monthly_sales.values

plt.figure(figsize=(10,5))
plt.plot(x, y, marker='o')
plt.title("Monthly Sales (Total Payment Value)")
plt.xlabel("Month")
plt.ylabel("Sales (R$)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Insights: Monthly Sales Trend

- **Strong growth:** Sales increased consistently from late 2016 through most of 2017, showing rapid growth in the e-commerce platform’s activity.  
- **Stabilization:** After mid-2017, sales values stabilized at a higher level, fluctuating around 1.4–1.6M R$.  
- **Outliers:** There is a noticeable sales spike in late 2017, which might indicate a seasonal event such as holiday promotions (e.g., Black Friday).  
- **Decline at the end:** Sales drop sharply in the last months of 2018, which is likely due to the dataset ending and not a real business decline.  
- **Business implication:** The platform was scaling quickly during the observed period, with significant sales volume and consistent customer demand.



## 2) Product Categories by Average Review Score

**Question:** Which product categories have the highest customer satisfaction?  
**Metric:** Average `review_score` per category, filtered to categories with a minimum number of reviews for robustness.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# prag minim de review-uri pentru a evita categoriile cu eșantion foarte mic
MIN_REVIEWS = 100

cat_stats = (
    df.groupby("product_category_name")
      .agg(
          avg_review=("review_score", "mean"),
          n_reviews=("review_score", "count")
      )
      .dropna(subset=["avg_review"])
)

cat_stats_filt = cat_stats[cat_stats["n_reviews"] >= MIN_REVIEWS]
top10 = cat_stats_filt.sort_values("avg_review", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.barh(top10.index.astype(str), top10["avg_review"])
plt.gca().invert_yaxis()
plt.title(f"Top 10 Categories by Average Review Score (n≥{MIN_REVIEWS})")
plt.xlabel("Average Review Score (1–5)")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()

top10
```

### Insights: Product Categories by Average Review Score

- **High satisfaction categories:** Media, books, and flowers categories have some of the highest average review scores (around 4.4–4.6).  
- **Customer experience:** These products are typically lower risk (low price, simple delivery), which may explain why customers rate them highly.  
- **Data quality note:** Scores are filtered to categories with at least 100 reviews to ensure reliability. Categories with fewer reviews might show inflated or biased averages.  
- **Business implication:** High-scoring categories indicate strong customer satisfaction. Focusing on these could help maintain loyalty, while further analysis is needed on lower-scoring categories to identify improvement areas.


## 3) Distribution of Review Scores

**Question:** How are customer review scores distributed?  
**Metric:** Frequency counts of `review_score` (1–5).


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

scores = df["review_score"].dropna().astype(int)

# bar chart pe scoruri discrete 1–5
freq = scores.value_counts().sort_index()

plt.figure(figsize=(7,4))
plt.bar(freq.index.astype(str), freq.values)
plt.title("Distribution of Review Scores")
plt.xlabel("Review Score")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# procentual (util pentru interpretare)
(freq / freq.sum()).round(3)
```

### Insights: Distribution of Review Scores

- **Overall satisfaction:** More than half of the reviews (≈56%) are 5 stars, showing that most customers are satisfied with their shopping experience.  
- **Positive reviews dominate:** If we combine 4- and 5-star reviews, nearly 75% of all ratings are positive.  
- **Negative experiences:** About 13% of reviews are 1 star, which is significant. This suggests that a non-negligible portion of customers had poor experiences (possibly late deliveries or product issues).  
- **Neutral reviews:** 3-star reviews make up about 8%, reflecting mixed customer satisfaction.  
- **Business implication:** While customer satisfaction is strong overall, the share of low reviews highlights areas to investigate (delivery times, product quality, or customer support).

```python
import numpy as np

# Define late deliveries
df["is_late"] = df["order_delivered_customer_date"] > df["order_estimated_delivery_date"]

# Drop rows where delivery dates are missing
review_data = df.dropna(subset=["order_delivered_customer_date", 
                                "order_estimated_delivery_date", 
                                "review_score"])

# Split groups
on_time = review_data.loc[review_data["is_late"] == False, "review_score"]
late = review_data.loc[review_data["is_late"] == True, "review_score"]

print("On-time mean review:", on_time.mean())
print("Late mean review:", late.mean())
```

```python
from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(on_time, late, equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

## Hypothesis Test: Late Deliveries and Review Scores

**Research Question:** Do late deliveries lead to lower customer review scores?

- **On-time deliveries average review score:** 4.21  
- **Late deliveries average review score:** 2.55  
- **T-statistic:** 92.22  
- **p-value:** < 0.001  

### Conclusion
We reject the null hypothesis (H₀).  
There is a **statistically significant difference** in review scores between on-time and late deliveries.  

- On-time deliveries receive **much higher review scores** (average > 4).  
- Late deliveries are strongly associated with **negative reviews** (average ≈ 2.5).  

### Business Implication
Delivery performance is a **key driver of customer satisfaction**.  
Improving logistics and reducing late deliveries could substantially lower the share of 1-star reviews and increase overall customer trust in the platform.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cleaned subset with 'is_late' and 'review_score'
review_data = df.dropna(subset=["order_delivered_customer_date", 
                                "order_estimated_delivery_date", 
                                "review_score"]).copy()

review_data.loc[:, "is_late"] = review_data["order_delivered_customer_date"] > review_data["order_estimated_delivery_date"]

plt.figure(figsize=(8,5))
sns.boxplot(x="is_late", y="review_score", data=review_data, hue="is_late", legend=False, palette="Set2")

plt.xticks([0,1], ["On-Time", "Late"])
plt.title("Review Scores: On-Time vs. Late Deliveries")
plt.xlabel("Delivery Status")
plt.ylabel("Review Score (1–5)")
plt.tight_layout()
plt.show()
```

### Insights: Review Scores – On-Time vs. Late Deliveries

- **Clear difference:** The boxplot shows that on-time deliveries typically receive higher review scores (median ≈ 5), while late deliveries have much lower scores (median ≈ 2).  
- **Distribution:**  
  - On-time orders are concentrated at 4–5 stars, with very few low ratings.  
  - Late deliveries show a wide spread of ratings, with many reviews at the minimum (1 star).  
- **Outliers:** There are a few low scores even for on-time orders, but they are rare compared to late deliveries.  
- **Business implication:** This visualization reinforces the hypothesis test results — late deliveries strongly correlate with poor customer satisfaction. Improving delivery reliability would likely reduce the number of negative reviews and increase overall ratings.
